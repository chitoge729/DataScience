```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
k = np.array([145,93,86,55])
my_labels = ["BTS", "SVT", "TXT", "SKZ"]
my_colors = ["purple", "gold", "royalblue", "maroon"]
plt.pie(k, colors = my_colors, startangle = 190)
plt.legend(labels = my_labels)
plt.show()
```
